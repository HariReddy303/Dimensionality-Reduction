```python
import pandas as pd
```

```python
df=pd.read_csv("CC GENERAL.csv")
df.head()
```

```python
df.shape
```

```python
df.drop(labels="CUST_ID",axis=1,inplace=True)
```

```python
df.info()
```

```python
df.fillna(method="ffill",inplace=True)
```

```python
df.info()
```

```python
type(df)
```

```python
df.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled=scaler.fit_transform(df)
scaled
```

```python
from sklearn.preprocessing import normalize
norm=normalize(scaled)
norm
```

```python
df1=pd.DataFrame(norm)
df1
```

```python
from sklearn.decomposition import PCA
pca=PCA()
pca.fit_transform(df1)
```

```python
df_pca=pd.DataFrame(pca.fit_transform(df1))
df_pca
```

```python
df2=df_pca.iloc[:,0:2]
df2.columns=["pc1","pc2"]
df2
```

```python
from sklearn.cluster import KMeans
means=KMeans(n_clusters=5)
means.fit_predict(df2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df2["pc1"],df2["pc2"])
plt.show()
```

```python
k=[2,3,4,5,6]
```

```python
silhouette_scores=[]
```

```python
from sklearn.metrics import silhouette_score
score=silhouette_score(df2,means.fit_predict(df2))
score
```

```python
silhouette_scores.append(score)
```

```python
plt.bar(k,silhouette_scores)
plt.show()
```
